# Problem 1: Unique Words in a Sentence


```python

# Step 1: Take input from user
sentence = input("Enter sentence: ")

# Step 2: Split sentence into words
words = sentence.split()

# Step 3: Convert list to set (removes duplicates)
unique_words = set(words)

# Step 4: Display results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Step 1: Find employee with highest salary
highest_employee = max(employees, key=employees.get)

# Step 2: Print result
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# Step 1: Initialize
max_value = numbers[0]
min_value = numbers[0]

# Step 2: Compare each number
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

# Step 3: Display result
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

# Problem 4: Count Products Above 1000

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

# Step 1: Count Present days
present_days = attendance.count("P")

# Step 2: Calculate percentage
percentage = (present_days / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

# Problem 6: Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

# Problem 7: Count Character Frequency

```python
text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```

# Problem 9: Check if a Key Exists in Dictionary

```python
employee = {
    "Ravi": 75000,
    "Anita": 68000
}

key_to_check = "Ravi"

if key_to_check in employee:
    print("Employee exists")
else:
    print("Employee does not exist")
```

# Problem 10: Calculate Average Marks


```python
marks = [70, 80, 65, 75]

# Step 1: Calculate sum
total = sum(marks)

# Step 2: Calculate average
average = total / len(marks)

print("Average Marks:", average)
```
